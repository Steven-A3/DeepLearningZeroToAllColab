### 환경설정

```
import tensorflow as tf
import pandas as pd
from tensorflow import keras
from tensorflow.keras import layers
print("Tensorflow version:", tf.__version__)

tf.random.set_seed(777) # for reproducibility
```

### 데이터

```
x_data = [1, 2, 3]
y_data = [1, 2, 3]
```

### Keras를 활용한 모델 구축

```
model = keras.Sequential()
model.add(layers.Dense(1, input_shape=[1]))  # x_data의 특성이 1개이므로 input_shape는 1로 지정
```

#### 확률적 경사하강법(SGD, Stochastic Gradient Descent) 사용

```
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
```

#### 모델 컴파일

```
model.compile(loss="mse", optimizer=optimizer, metrics=['mse'])
```

#### 결과확인

```
history = model.fit(x_data, y_data, verbose=1, epochs=1000) # verbose를 1로 지정하면 모델 훈련 과정을 출력
```

```
hist = pd.DataFrame(history.history)
hist["epoch"] = history.epoch
hist.tail()
```

#### 가중치와 절편값 확인

```
W, b = model.get_weights()
print("Weight:", W, "Bias:", b)
```
