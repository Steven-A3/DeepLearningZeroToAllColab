### 환경설정

```
import tensorflow as tf
import pandas as pd
import numpy as np

from tensorflow import keras
from tensorflow.keras import layers

print("Tensorflow version:", tf.__version__)

tf.random.set_seed(777)  # for reproducibility
```

```
from google.colab import drive
drive.mount("/content/gdrive")
```

```
xy = np.loadtxt('/content/gdrive/My Drive/Colab Notebooks/data/data-03-diabetes.csv', delimiter=',', dtype=np.float32)
xy
```

```
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]

print(x_data.shape, y_data.shape)  # 759개의 데이터, 데이터의 특성은 8개
```

### Keras를 활용한 모델 구축

```
model = keras.Sequential()
model.add(layers.Dense(12, input_dim=8 , activation="sigmoid")) # x_data의 특성 개수가 8개 이므로 input_dim은 8
model.add(layers.Dense(8, activation="sigmoid"))
model.add(layers.Dense(1, activation="softmax"))
```

#### 확률적 경사하강법(SGD, Stochastic Gradient Descent) 사용

```
optimizer = tf.keras.optimizers.SGD(learning_rate=1e-5)
```

#### 모델 컴파일

```
model.compile(loss="binary_crossentropy", optimizer=optimizer, metrics=["acc"])
```

#### 모델 훈련

```
model.fit(x_data, y_data, verbose=0, epochs=10000)
```

#### 모델 평가

```
_, accuracy = model.evaluate(x_data, y_data)
print("Accuracy:", accuracy)
```
